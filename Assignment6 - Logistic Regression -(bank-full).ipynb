```python
#spliting into contineous and categorical

X1 = df[df.columns[[0,5,9,11,12,13,14]]]
X1

```

```python
X2 = df[df.columns[[1,2,3,4,6,7,8,10,15,16]]]
X2
```

```python
# Data Transformation 
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
for col in X2:
    # Create Object of LabelEncoder
    LE = LabelEncoder()
    X2[col] = LE.fit_transform(X2[col])
X2.dtypes
```

```python
X = pd.concat([X1,X2], axis = 1)
X.head()
```

```python
# Split dataset in input and output
x = X.drop('y',axis=1)  # input
y1 = X['y']             # output
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss_x = ss.fit_transform(x)
```

```python
# Train and Test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y1,test_size=0.3)
```

```python
# Model Fitting
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, Y_train)
Y_pred_train = LR.predict(X_train)
Y_pred_test = LR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, log_loss
ac_1 = accuracy_score(Y_train, Y_pred_train)
ac_2 = accuracy_score(Y_test, Y_pred_test)
print("accuracy_score_1: ", ac_1)
print("accuracy_score_2: ", ac_2)
```

```python
fs_1 = f1_score(Y_train, Y_pred_train)
fs_2 = f1_score(Y_test, Y_pred_test)
print("f1_score_1: ", fs_1)
print("f1_score_2: ", fs_2)
```

```python
ll_1 = log_loss(Y_train, Y_pred_train)
ll_2 = log_loss(Y_test, Y_pred_test)
print("Log_Loss_1: ", ll_1)
print("Log_Loss_2: ", ll_2)
```

```python
# cross validation method
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
from sklearn.model_selection import train_test_split

training_accuracy = []
test_accuracy = []
for i in range(1,101):
    X_train,X_test,Y_train,Y_test = train_test_split(x,y1,test_size=0.30,random_state=i)
    logreg.fit(X_train,Y_train)
    Y_pred_train = logreg.predict(X_train)
    Y_pred_test = logreg.predict(X_test)
    training_accuracy.append(accuracy_score(Y_train,Y_pred_train))
    test_accuracy.append(accuracy_score(Y_test,Y_pred_test))
   
print("Cross validation Training score: ", np.mean(training_accuracy).round(2))
print("Cross validation Test score: ", np.mean(test_accuracy).round(2))

## Cross validation Training score:  0.89
## Cross validation Test score:  0.89
```

```python
# KFOLD CROSS VALIDATION
from sklearn.model_selection import KFold
KFold = KFold(n_splits=5)

training_accuracy = []
test_accuracy = []
for train_index,test_index in KFold.split(x):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = y1[train_index], y1[test_index]
    logreg.fit(X_train,Y_train)
    Y_pred_train = logreg.predict(X_train)
    Y_pred_test  = logreg.predict(X_test)
    training_accuracy.append(accuracy_score(Y_train,Y_pred_train))
    test_accuracy.append(accuracy_score(Y_test,Y_pred_test))
    
print("Kfold- Cross validation Training score: ", np.mean(training_accuracy).round(2))
print("Kfold- Cross validation Test score: ", np.mean(test_accuracy).round(2))
```

```python
df["Y_prob"] = logreg.predict_proba(X)[:,1]
```
